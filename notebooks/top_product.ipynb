```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Top 10 products

```python
products = pd.read_csv('../data/processed/cleaned_sales_data.csv', parse_dates=['InvoiceDate'])

products['Revenue'] = products['Quantity'] * products['UnitPrice']
top_products = products.groupby('Description')['Revenue'].sum().sort_values(ascending=False).head(10)
total_revenue = products['Revenue'].sum()
revenue_share_percent = ((top_products / total_revenue) * 100).round(2)
```

```python
plt.figure()

plt.barh(top_products.index, revenue_share_percent.values)

plt.title('Top 10 Products by Revenue Share (%)')
plt.xlabel('Share %')

plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig('../outputs/graphs/top_products_share.png')

plt.show()
```
